# Matrix Multiplication Using Numpy Library

# Scalar / Dot Product

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import time

# 100* 100

start_time_1=time.time()
arr1=np.random.randint(100,size=(100,100)) #create a 100*100 array ranging from 0 to 99

print("arr1=", arr1)

arr2=np.random.randint(100,size=(100,100)) #create another 100*100 array ranging from 0 to 99

print("\narr2=", arr2)




dot1=np.dot(arr1,arr2)
print("\nFinal Result Array=" ,dot1)

time1=time.time()-start_time_1

print("\nTime to multiply  2 100*100 matrices", time1,"seconds")




```
